```python
import pandas as pd

prices = [
	(1, 2.12),
	(2, 2.56),
	(3, 3.10),
	(4, 3.16),
	(5, 3.58),
	(6, 5.12),
	(7, 5.16),
	(8, 5.20),
	(9, 4.12),
	(10, 4.10),
	(11, 3.65),
	(12, 4.25),
]
df = pd.DataFrame(prices, columns = ['month', 'price in PLN'])
df = df.set_index('month')
df['price in USD'] = df['price in PLN'] / 4
df['price in USD'].plot(ylim=(0, 2), kind='line', grid=True, linestyle='dashed', color='red', linewidth=3, title="Prices of goods (USD)")
df
```
